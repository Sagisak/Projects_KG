```python
import random
import pandas as pd
import os

# Set the seed for reproducibility
random.seed(42)

# Product categories
produk_sering_beli = [
    "Novel Fiksi",
    "Buku Non-Fiksi (Pengembangan Diri, Bisnis, dll.)",
    "Buku Pelajaran/Edukasi (Termasuk Latihan Soal, Kamus)",
    "Komik/Novel Grafis",
    "Buku Anak",
    "Perlengkapan Seni/Kreatif",
    "Alat Musik",
    "Alat Tulis Umum (Kantor/Sekolah)",
    "Jurnal/Perencana (Planner)",
    "Perlengkapan Kantor atau Sekolah",
    "Mainan/Collectible item",
    "Alat Olahraga"
]

# Purchase motivations
motivasi_pembelian = [
    "Menambah pengetahuan",
    "Hiburan atau relaksasi",
    "Kebutuhan pekerjaan/sekolah (alat tulis, perlengkapan kantor)",
    "Mendukung hobi (misalnya, melukis, dekorasi, musik, lainnya)",
    "Kado atau hadiah",
    "Untuk mengekspresikan kreativitas",
    "Sebagai bentuk 'self-reward' atau apresiasi diri"
]

# Shopping channels
saluran_belanja = [
    "Toko Gramedia",
    "Website Gramedia.com",
    "Aplikasi Gramedia.com",
    "Aplikasi Gramedia Digital (ebook)",
    "Marketplace (Shopee, Tokopedia, dll.)",
    "Toko buku independen lainnya",
    "Toko alat tulis lainnya"
]

hobi = [
    "Membaca buku atau majalah ",
    "Menulis (diari, blog, atau lainnya)",
    "Melukis atau menggambar ",
    "Bermain musik",
    "Bermain game (offline/online)",
    "Olahraga",
    "Menghadiri acara budaya atau edukasi",
    "Belanja online/offline",
    "Travelling"
]

minat = [
    "Buku fiksi (novel, komik, buku anak)",
    "Buku non-fiksi (pengembangan diri, sejarah, agama)",
    "Buku Edukasi (Buku Teks, Latihan Soal, Kamus)",
    "Alat Musik & kesenian",
    "Mainan",
    "Alat Olahraga",
    "Alat Tulis kantor atau sekolah",
    "Kesehatan dan Kecantikan"
]

likert_responses = [
    "Sangat Tidak Setuju",
    "Tidak Setuju",
    "Netral",
    "Setuju",
    "Sangat Setuju"
]

def generate_data(num_rows=1000):
    data = {
        "Produk sering beli": [
            "; ".join(random.sample(produk_sering_beli, k=random.randint(1, 3)))
            for _ in range(num_rows)
        ],
        "Saluran belanja": 
            [random.choice(saluran_belanja) for _ in range(num_rows)],

        "Motivasi utama pembelian": [
            "; ".join(random.sample(motivasi_pembelian, k=random.randint(1, 3)))
            for _ in range(num_rows)
        ],
        "Hobi": [
            "; ".join(random.sample(hobi, k=random.randint(1, 3)))
            for _ in range(num_rows)
        ],
        "Minat": [
            "; ".join(random.sample(minat, k=random.randint(1, 3)))
            for _ in range(num_rows)
        ],
         "Pernyataan: Membaca adalah bagian penting dari pengembangan diri saya": [
            random.choice(likert_responses) for _ in range(num_rows)
        ],
        "Pernyataan: Saya sering mencari inspirasi untuk kegiatan kreatif dari buku atau alat tulis": [
            random.choice(likert_responses) for _ in range(num_rows)
        ],
        "Pernyataan: Saya berusaha menerapkan gaya hidup minimalis dalam hal kepemilikan barang": [
            random.choice(likert_responses) for _ in range(num_rows)
        ],
        "Pernyataan: Saya nyaman menggunakan teknologi digital untuk membaca dan mencatat": [
            random.choice(likert_responses) for _ in range(num_rows)
        ] 
        
    }
    return pd.DataFrame(data)

def get_available_filename(base_name="DataDummyPsikografi_15000", ext="csv"):
    version = 1
    folder = "DataDummyPsikografi_15000"
    os.makedirs(folder, exist_ok=True)  # Make sure the folder exists
    while True:
        filename = f"{base_name}_v{version}.{ext}"
        full_path = os.path.join(folder, filename)
        if not os.path.exists(full_path):
            return full_path  # Return full path including folder
        version += 1

def main(num_rows=1000):
    df = generate_data(num_rows)  # Assumes this function is defined elsewhere
    filepath = get_available_filename()  # This now includes the folder
    df.to_csv(filepath, index=False, encoding="utf-8")
    print(f"Data saved as {filepath}")

if __name__ == "__main__":
    # Change the number below to adjust number of rows
    main(num_rows=50000)
```
