import pandas as pd
from fuzzywuzzy import process

def clean_product_names(judulbook_filtered, Product, threshold=80):
    
    unique_products = set() 
    mapping = {}

    for product in judulbook_filtered['Product'].dropna().unique():
        if not isinstance(product, str):
            continue

        result = process.extractOne(product, list(unique_products)) if unique_products else None

        if result:
            match, score = result
            if score >= threshold:
                mapping[product] = match
            else:
                unique_products.add(product)
                mapping[product] = product
        else:
            unique_products.add(product)
            mapping[product] = product

    # **Update kolom 'Product' di DataFrame**
    judulbook_filtered["Product"] = judulbook_filtered["Product"].apply(lambda x: mapping.get(x, x))
    
    return judulbook_filtered

# **Panggil fungsi dengan dataset yang sudah kamu miliki**
judulbook_filtered = clean_product_names(judulbook_filtered, "Product")
