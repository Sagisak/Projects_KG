```python
import pandas as pd
shopee_data = pd.read_csv('judulbook_filtered1.csv')
shopee_data
```

```python
#Normalization for Product Column (Deleting [],!/\|)
import re

def normalize_title(title):
    title = re.sub(r"[\[\(].*?[\]\)]", "", title)
    
    title = re.sub(r"[^\w\s]", "", title)
    
    title = title.lower()
    
    title = " ".join(title.split())

    return title

shopee_data["product_without_signs"] = shopee_data["Product"].apply(normalize_title)
```
